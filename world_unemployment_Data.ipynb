Q1. What is the average global unemployment rate for the latest year available in the dataset?



```python
# Here we are fetching the average of of our Data column name '2022',however we find mean but we will not plot this because it is asked
#over all average global unemployment rate for the latest year available in the dataset,so we get only single value for all the rows.
average_of_global_unemployment=df['2022'].mean()
average_of_global_unemployment
```


Q2. Which country had the highest unemployment rate in 2022?



```python
# Here we are fetching top 10 country who had the highest unemployment rate in 2022 by fetching
#Data with pre-defined function nlargest and we are giving desired row value and column name.
top_10_unemployment_2022=df.nlargest(10,'2022')[['Country Name','2022']]
top_10_unemployment_2022
```

```python
# as we have imported the library seaborn in first cell, to get more effects in our visualization we call the fuction 'pallete' from sns(seaborn),
palette=sns.color_palette('inferno',len(top_10_unemployment_2022))

# Plot a bar chart with rotated country names and Seaborn color palette
plt.figure(figsize=(10,6))
bars=plt.bar(top_10_unemployment_2022['Country Name'],top_10_unemployment_2022['2022'],color=palette)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45,ha='right')

# Add title
plt.xlabel('country')
plt.ylabel('top_10_unemployment_rate_2022')
plt.title('Top 10 Countries with Highest Unemployment Rates in 2022')

# To display the valueson each bar
for bar, color in zip(bars,palette):
  yval=bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,yval,round(yval,2),ha='center',va='bottom')
plt.show()
```


Q3. How has the global average unemployment rate changed from 1998 to 2021?


```python
# Here We are average of numeric columns as we have asked by using loc function and giving range all row and starts with '1998' and end at '2022'.
average_of_global_unemployment_1998_2021=df.loc[:,'1998':'2022'].mean()
```

```python
palette=sns.color_palette("bright")[0]

plt.figure(figsize=(12,4))
plt.style.use('dark_background')
line=plt.plot(average_of_global_unemployment_1998_2021.index,average_of_global_unemployment_1998_2021,color=palette,marker='o')
# Here we  are setting the value on plot
for x, y in zip(average_of_global_unemployment_1998_2021.index, average_of_global_unemployment_1998_2021):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', color='skyblue')

plt.xticks(rotation=45,ha='right')
plt.xticks(color='skyblue')
plt.yticks(color='skyblue')
plt.ylabel('average_of_global_unemployment',color='skyblue')
plt.xlabel('Years',color='skyblue')
plt.title('average_of_global_unemployment_1998_2021',color='skyblue')
# Here we are setting the border of the plot
for spine in plt.gca().spines.values():
  spine.set_edgecolor('skyblue')
  spine.set_linewidth(2)
  spine.set_linestyle('solid')
```


Q4. Which country has shown the most improvement in reducing unemployment over the last decade?


```python
# Here we are fetching the country name and numeric  columns.
dataframe=df[['Country Name'] + [str(year) for year in range(2012,2022+1)]]

# As we have to make difference from last decade so we fetching the difference between last year(2012) of that decade and this current year(2022)column wise.
change_in_unemployement=dataframe.set_index('Country Name').diff(axis=1)

# we are  adding the value colunm wise.
sum_of_unemployement=change_in_unemployement.sum(axis=1)

# Here we are fetching the maxid of that country who is having most improvement in reducing unemployment over the last decade.
most_improved_country=sum_of_unemployement.idxmin()

# here we are fetching that particular row where that country exists.
most_improved_value=sum_of_unemployement.loc[most_improved_country]

 # now its time to print that country and value which present actual improvement points.
print('The name of the country is',most_improved_country)
print('The name of the value is',most_improved_value)
```


Q5. Is there a noticeable impact of the 2008 financial crisis on global unemployment?



```python
# Here we are fetching near by year of 2008, we can come across the impact of 2008 financial crisis on global umemployment.
data_of_2008=df[[ str(year)for year in range (2007,2011)]]
avg=data_of_2008.mean(axis=0).round(2)
```

```python
plt.figure(figsize=(10,4))
avg.plot(style='o-', markerfacecolor='red', markersize=7, linewidth=2,color='skyblue',ylim=[6,9])
for i, value in enumerate(avg.values):
    plt.text(i, value, f'{value:.2f}', ha='left', va='bottom', color='red')

plt.ylabel('Average impact of 2008',color='skyblue')
plt.xlabel('Years',color='skyblue')
plt.title('Unemployment impact rate 2008 world wide',color='skyblue')
plt.xticks(color='skyblue')
plt.yticks (color='skyblue')
for spine in plt.gca().spines.values():
  spine.set_edgecolor('skyblue')
  spine.set_linewidth(2)
  spine.set_ls('solid')

#Yes, the 2008 financial crisis has had an effect on the unemployment rate worldwide. as impact can be oserved through this visualization.
```


Q6. How did the COVID-19 pandemic affect unemployment rates in 2020 and 2021?


```python
unemployment_rates_in_2020_2021=df[[str(year)for year in range(2019,2022+1)]]
avg=unemployment_rates_in_2020_2021.mean(axis=0).round(2)
avg
```

```python
plt.figure(figsize=(10,4))
avg.plot(marker='o',markerfacecolor='red',markersize=7,c='skyblue',ylim=[6,9])
for i,value in enumerate(avg.values):
  plt.text(i,value,f'{value:.2f}',ha='left',va='bottom',c='skyblue')
plt.ylabel('Impact of covid-19 on unemployment',c='skyblue')
plt.xlabel('year',c='skyblue')
plt.title('COVID-19 pandemic affect unemployment rates in 2020 and 2021',c='skyblue')
plt.yticks(c='skyblue')
plt.xticks(c='skyblue')
for spine in plt.gca().spines.values():
    spine.set_edgecolor('skyblue')
    spine.set_linewidth(2)
    spine.set_linestyle('solid')
plt.show()

```


Q7. What are the top 5 countries with the lowest average unemployment rate from 1998 to 2022?



```python
selected_data = df[['Country Name'] + [str(year) for year in range (1998,2022+1)]]
a=selected_data.apply(pd.to_numeric, errors='coerce')
average_unemployment = a.groupby(df['Country Name'])[[str(year) for year in range (1998,2022+1)]].mean()
top_5_lowest_unemployement=average_unemployment.mean(axis=1).nsmallest(5)
top_5_lowest_unemployement
```

```python
 palette =sns.color_palette('inferno',len(top_5_lowest_unemployement))
 plt.figure(figsize=(10,4))
 bar=plt.bar(top_5_lowest_unemployement.index,top_5_lowest_unemployement.values,color=palette, edgecolor='skyblue')
 plt.ylim=(0,6)
 plt.xticks(rotation=0,ha='right',color='skyblue')
 plt.yticks(color='skyblue')
 plt.xlabel('Top 5 counteries',color='skyblue', fontsize=10)
 plt.ylabel('Average',color='skyblue')
 plt.title('5 countries with the lowest average unemployment rate from 1998 to 2022',color='skyblue')
for bar, color in zip(bar,palette):
     yval=bar.get_height()
     plt.text(bar.get_x()+bar.get_width()/2,yval,round(yval,2),ha='center',va='bottom',color='skyblue')

for spine in plt.gca().spines.values():
    spine.set_edgecolor('skyblue')
    spine.set_linewidth(2)
    spine.set_linestyle('solid')




plt.show()
```

Q9. Which region has the most stable employment market over the years?

```python


unemployment_2022 = df['2022']
unemployment_2022 = unemployment_2022.dropna()
plt.figure(figsize=(10, 4))
ax=sns.histplot(unemployment_2022, bins=30, kde=True, color='grey',edgecolor='skyblue')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', color='white', fontsize=8)
plt.title('Histogram of Unemployment Rates for 2022',color='skyblue')
plt.xlabel('Unemployment Rate',color='skyblue')
plt.ylabel('Frequency',color='skyblue')
plt.yticks(color='skyblue')
plt.xticks(color='skyblue')
plt.gca().set_ylim(0, 26)
ax.grid(True)
for spine in plt.gca().spines.values():
    spine.set_edgecolor('skyblue')
    spine.set_linewidth(2)
    spine.set_linestyle('solid')
plt.show()
```
